```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
features_path = "../data/features.csv"
features_df = pd.read_csv(features_path)
features_df = features_df.drop(columns=features_df.columns[:20])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
features_df.boxplot(ax=ax)
ax.set_xticklabels(features_df.columns, rotation=45)
ax.set_ylabel("Feature Values")
ax.set_xlabel("Features")
plt.tight_layout()
plt.savefig('figures/ascertian_features_box_plots.png')
plt.show()
```
