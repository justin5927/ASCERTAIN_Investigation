```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
features_path = "../data/features.csv"
features_df = pd.read_csv(features_path)
features_df = features_df.drop(columns=features_df.columns[:19])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
features_df.boxplot(ax=ax)
ax.set_xticklabels(features_df.columns, rotation=45)
ax.set_title("Box Plot of ECG Signal Features")
ax.set_ylabel("Feature Values")
plt.tight_layout()
plt.show()
```
