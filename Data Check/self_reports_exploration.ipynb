```python
# Imports 
import scipy.io
import matplotlib.pyplot as plt
```

```python
ecg_path = '../ASCERTAIN_Features/Dt_SelfReports.mat'
mat = scipy.io.loadmat(ecg_path)
```

```python
# Key Name
key_names = list(mat.keys())
print(key_names)
key = list(key_names)[5]
```

```python
# Sizes of the data
print(f"Number of Rating: {len(mat[key])}")
print(f"Number of Participants: {len(mat[key][0])}")
print(f"Number of Recording: {len(mat[key][0][0])}")
```

```python
rating_num = len(mat[key])
participants_num = len(mat[key][0])
recording_num = len(mat[key][0][0])
emotions = ["Arousal", "Valence", "Engagement", "Liking", "Familiarity"]
emo_dict = {emo: [] for emo in emotions}
```

```python
features = []
for rate in range(rating_num):
    for par in range(participants_num):
        for record in range(recording_num):
            emo_dict[emotions[rate]].append(mat[key][rate][par][record])
```

```python
emo = emotions[4]
plt.hist(emo_dict[emo])
plt.title(f"Histogram of {emo}")
plt.grid()
plt.xlabel("Ratings")
plt.ylabel("Counts")
```
