```python
# Imports 
import scipy.io
import matplotlib.pyplot as plt
```

```python
ecg_path = '../ASCERTAIN_Features/Dt_ECGFeatures.mat'
mat = scipy.io.loadmat(ecg_path)
```

```python
# Key Name
key_names = list(mat.keys())
print(key_names)
key = list(key_names)[4]
```

```python
# Sizes of the data
print(f"Number of Participants: {len(mat[key][0])}")
print(f"Number of Recordings: {len(mat[key][0][0])}")
print(f"Number of ECG Features: {len(mat[key][0][0][0])}")
```

```python
participants_num = len(mat[key][0])
recording_num = len(mat[key][0][0])
ecg_features_num = len(mat[key][0][0][0])
```

Features for reference


Grabbing one feature

```python
features = []
for p in range(participants_num):
    for r in range(recording_num):
        features.append(mat[key][0][p][r][27])
```

```python
plt.hist(features)
plt.title(f"Histogram of Stats Measurements over HRV")
plt.grid()
plt.xlabel("Scores")
plt.ylabel("Counts")
```
